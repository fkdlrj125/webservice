# 결정 트리 모델 시각화

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

## 학습 데이터, 테스트 데이터 셋 분리

```python
iris = load_iris()
feature = iris.data
target = iris.target

x_train, x_test, y_train, y_test = train_test_split(
    feature,
    target,
    test_size=0.2,
    random_state=11,
)
```

## 하이퍼 파라미터 조정 안 한 결정트리객체 생성

```python
# 결정트리 객체 생성
# 하이퍼 파라미터를 디폴트로 설정하면 분류를 가능한 최대로 하기 때문에 overfitting이 발생할 수 있음
dt_clf = DecisionTreeClassifier(random_state=11)

# 학습
dt_clf.fit(x_train, y_train)

# 예측
pred = dt_clf.predict(x_test)

# 평가
acc = accuracy_score(y_test, pred)
f'테스트 데이터 세트 정확도 : {acc}'
```

```python
# 모델의 깊이 확인
dt_clf.get_depth()
```

```python
# 모델의 리프 노드 개수
dt_clf.get_n_leaves()
```

# 모델 시각화

```python
from sklearn.tree import export_graphviz
```

```python
import graphviz
```

```python
export_graphviz(
    dt_clf,
    out_file='tree.dot', # dot파일로 그래프 저장
    class_names=iris.target_names, # label데이터에 해당하는 이름
    feature_names=iris.feature_names, # feature명
    filled=True, # 노드에 자동으로 색 설정
)
```

```python
iris.feature_names
```

- 모델이 학습을 할 때 중요하게 판단할 속성(feature)를 지정하지 않으면 알아서 설정
    - 스스로 설정한 속성이 최선의 판단 속성이라고 보장하진 못 함
<BR><BR>
- 모델이 분류를 할 때 균일한 데이터를 만들기 위해 최대한 노력
    - 균일한 데이터 : 유사한 데이터끼리 모여 있는 데이터
    - 완벽히 균일하게 데이터를 분류하면 overfitting이 발생
<BR><BR>
- 균일도 기반 규칙 조건
    - 분류를 처음할 땐 가장 균일한 데이터를 기준으로 분류
<BR><BR>
- 균일도 측정 방법
    - 정보 이득(IG : Information Gain)
        - 엔트로피(entropy) 기반
            - 엔트로피 
                - 데이터의 혼잡도=불확실성
                - 다른 값이 섞여 있으면 엔트로피가 높음
                    - 정보가 빈번하지 않고 예측 불가능하면 정보 효율성이 높음 : 희귀한 정보
                    - 데이터 분류에 분리
                - 같은 값이 섞여 있으면 엔트로피 낮음
                    - 정보가 빈번하고 예측 가능하면 정보 효율성이 낮음 : 흔한 정보
                    - 데이터 분류에 유리
<BR><BR>
        - 정보이득 = 1-엔트로피 지수
        - Decision Tree는 IG값이 높은 노드(속성)를 분할
            - 데이터 혼잡도가 낮은 데이터, 균일도가 높은 데이터를 분할
<BR><BR>
    - 지니 계수(default) : 데이터의 균일도(균형도)를 측정
        - 0에 가까울 수록 균일도가 높음 : 균일한 데이터

```python
with open('tree.dot') as file:
    graph = file.read()
    
graphviz.Source(graph)
# 하이퍼 파라미터를 설정하지(튜닝) 않아 최대한 분류한 모델
# gini : 지니 계수(0에 가까울 수록 균일도가 높음)
# samples : 데이터 개수
# value : 분류한 데이터의 개수
# class : 분류할 target(종 이름)
```

```python
## 이미지 파일로 저장
g_s = graphviz.Source(graph)
g_s.render(filename='iris_파라미터 조정하지 않음', format='png') # 파일명과 확장자를 따로 설정
# dot파일과 이미지파일 2개 생성
```

# 하이퍼 파라미터 튜닝

##  max_depth=3

```python
# 결정트리 객체 생성
dt_clf = DecisionTreeClassifier(
    random_state=11, 
    max_depth=3,
)

# 학습
dt_clf.fit(x_train, y_train)

# 예측
pred = dt_clf.predict(x_test)

# 평가
acc = accuracy_score(y_test, pred)
f'테스트 데이터 세트 정확도 : {acc : .4f}'
```

```python
export_graphviz(
    dt_clf,
    out_file='tree_max_depth_3.dot', # dot파일로 그래프 저장
    class_names=iris.target_names, # label데이터에 해당하는 이름
    feature_names=iris.feature_names, # feature명
    filled=True, # 노드에 자동으로 색 설정
)

with open('tree_max_depth_3.dot') as file:
    graph = file.read()
    
graphviz.Source(graph)
# 하이퍼 파라미터를 설정하지(튜닝) 않아 최대한 분류한 모델
# gini : 지니 계수(0에 가까울 수록 균일도가 높음)
# samples : 데이터 개수
# value : 분류한 데이터의 개수
# class : 분류할 target(종 이름)
```

```python
dt_clf.get_depth()
```

```python
dt_clf.get_n_leaves()
```

## min_samples_leaf = 4
- 리프노드가 되기 위해 필요한 최소한의 샘플 데이터 수
- 분류를 할 때 분류된 리프노드의 데이터 수가 설정 값보다 낮다면 분류하지 않음

```python
dt_clf = DecisionTreeClassifier(
    random_state=11, 
    min_samples_leaf=4
)

dt_clf.fit(x_train, y_train)

pred = dt_clf.predict(x_test)

acc = accuracy_score(y_test, pred)
f'테스트 데이터 세트 정확도 : {acc : .4f}'
```

```python
export_graphviz(
    dt_clf,
    out_file='tree_min_samples_leaf_4.dot', # dot파일로 그래프 저장
    class_names=iris.target_names, # label데이터에 해당하는 이름
    feature_names=iris.feature_names, # feature명
    filled=True, # 노드에 자동으로 색 설정
)

with open('tree_min_samples_leaf_4.dot') as file:
    graph = file.read()
    
graphviz.Source(graph)
# 하이퍼 파라미터를 설정하지(튜닝) 않아 최대한 분류한 모델
# gini : 지니 계수(0에 가까울 수록 균일도가 높음)
# samples : 데이터 개수
# value : 분류한 데이터의 개수
# class : 분류할 target(종 이름)
```
