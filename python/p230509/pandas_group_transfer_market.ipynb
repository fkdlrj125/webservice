```python
# [문제] 국가별 선수 수
group_nation = market.groupby('nation')

group_nation.size().sort_values(ascending=False).head()
```

```python
# [문제] 국가별 선수의 이적료 합
# group_nation.value.sum()
# group_nation.sum(numeric_only=True).value
group_nation.value.sum().sort_values(ascending=False).head()
```

```python
# [문제] 국가별 선수의 이적료 평균 (소수 2자리까지 표시)
# group_nation.value.mean().round(2)
group_nation.value.mean().round(2).sort_values(ascending=False).head()
```

```python
# 국가별 나이별 선수 수
group_nation_age = market.groupby(['nation', 'age'])
```

```python
group_nation_age.size()
```

```python
# 아르헨티나, 22세 데이터프레임 추출 - 키가 튜플이기 때문에 튜플로만 추출가능
group_nation_age.get_group(('Argentina', 22))
```

```python
# 인덱스
# group_nation_age.size().index
group_nation_age.indices #[]안에 있는 값은 행의 인덱스
```

```python
type(group_nation_age.indices)
```

```python
# 멀티인덱스 : 인덱스가 2개 이상인 것
# 기초 통계량
desc = group_nation_age.describe()
```

```python
# 행의 멀티인덱스
desc.index
```

```python
# 열의 멀티인덱스
desc.columns
```

```python
# number 컬럼 추출
desc.number
```

```python
# Argentina행 추출
# desc[0:1] # 슬라이싱으로 잘 추출되지 않음
# desc.loc['Argentina']

# Argentina행의 age가 25인 행 추출
# desc.loc['Argentina'].loc[25]
desc.loc[('Argentina', 25)]
```

```python
# number 컬럼의 mean(평균) 추출
# desc.number.mean # mean 메소드가 있기 때문에 via로 접근 불가능
# desc['number']['mean']
desc[('number', 'mean')]
```

```python
# value 컬럼 추출
desc.value
```
