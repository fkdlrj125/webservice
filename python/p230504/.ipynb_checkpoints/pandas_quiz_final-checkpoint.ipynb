```python
# 네이버 증권에서 국내 증시에 인기 검색 종목 테이블 추출 후 
# 모든 값이 결측치인 행만 삭제
# 타입 알맞게 변경
# 국내증시.csv 파일로 저장(인덱스 제외)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from time import sleep
import pandas
import os
import numpy
```

```python
# 크롬 브라우저 최대화 설정
chromeOptions = webdriver.ChromeOptions()
# chromeOptions.add_argument('headless')
chromeOptions.add_argument('--start-maximized')

# 크롬 브라우저 실행
driver = webdriver.Chrome(options=chromeOptions)

# 1. 네이버 접속
driver.get('https://www.naver.com')

sleep(1)
# 2. 증권 탭 클릭
finance_xpath = '//*[@id="NM_FAVORITE"]/div[1]/ul[2]/li[3]/a'
driver.find_element(By.XPATH, finance_xpath).click()

sleep(1)
# 3. 국내 증시 탭 클릭
sise_xpath = '//*[@id="menu"]/ul/li[2]/a'
driver.find_element(By.XPATH, sise_xpath).click()

sleep(1)
# 4. 인기 검색 종목 더보기 클릭
popular_xpath = '//*[@class="box_top_submain2"]/div[2]/a'
driver.find_element(By.XPATH, popular_xpath).click()

sleep(1)

# 5-1. 페이지 소스를 통해 데이터프레임 생성
df = pandas.read_html(driver.page_source)[1]

# # 5-2. 테이블의 th, td를 추출 후 데이터프레임 생성
# # 테이블 추출
# table_element = driver.find_element(By.CSS_SELECTOR, 'table.type_5')

# # tr태그 추출
# table_rows = table_element.find_elements(By.TAG_NAME, 'tr')
# table_list = []

# # tr태그 내 th, td 태그 추출
# for row in table_rows:
#     row_elements = row.find_elements(By.CSS_SELECTOR, 'th, td')
#     elements=[]

#     for element in row_elements:
#         elements.append(element.text)

#     table_list.append(elements)

# # 추출한 th태그를 column명으로 저장하고 td는 row로 저장    
# df = pandas.DataFrame(table_list[1:], columns=table_list[0])

# # 순위의 빈 문자열을 결측치로 변경
# df.loc[df.순위 == '','순위'] = numpy.NaN

# # 행의 모든 데이터가 결측치인 행 삭제
# df.dropna(how='all')

# [전처리 작업] 
# 6. 행의 모든 데이터가 결측치인 행 삭제 후 원본에 적용
df.dropna(how='all', inplace=True)

# 7-1. 컬럼의 Dtype 변경
# df = df.astype({'순위' : 'int16', '현재가' : int, '전일비' : int, '거래량' : int, '시가' : int, '고가' : int, '저가' : int})

# 7-2. 마스크를 통한 컬럼명 추출 후 Dtype 변경
change_columns = df.loc[:, df.dtypes=='float64'].columns[:-2]
for column in change_columns:
    df[column] = df[column].astype(int)

# 인덱스 재설정
df.reset_index(drop=True, inplace=True)
# [전처리 작업 끝]

# 8. 데이터 프레임을 .csv파일로 저장
df.to_csv('국내증시.csv', encoding='UTF-8' ,index=False)

sleep(1)
# 9. 브라우저 종료
driver.close()
```
