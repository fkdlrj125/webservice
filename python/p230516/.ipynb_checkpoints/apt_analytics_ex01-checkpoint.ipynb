```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# 파일 읽어오기
apt_price = pd.read_csv('../p230515/apt_price.csv')
apt_price.head()

plt.rc('font', family='D2coding')
```

```python
# 연도별 분양가격의 평균을 박스그래프, 바이올린그래프로 시각화
# 박스
sns.boxplot(
    data=apt_price,
    x='연도',
    y='분양가격',
).figure.set_size_inches(10,5)

plt.grid(axis='y', alpha=0.8)
```

```python
g_year_price_mean = apt_price.groupby('연도').분양가격
g_year_price_mean.get_group(2015).to_frame().boxplot()
```

```python
# 바이올린
sns.violinplot(
    data=apt_price,
    x='연도',
    y='분양가격',
).figure.set_size_inches(10,5)

plt.grid(axis='y', alpha=0.8)
```

```python
plt.figure(
    figsize=(8,6)
)

sns.lineplot(
    data=apt_price,
    x='연도',
    y='분양가격',
    hue='지역명',
    errorbar=None,
    estimator=np.mean
)

plt.legend(
    loc=(1.05 , 0.25)
)
```

```python
plt.figure(
    figsize=(8,6)
)

fig = sns.lineplot(
    data=apt_price,
    x='연도',
    y='분양가격',
    hue='지역명',
    errorbar=None,
    estimator=np.mean
)

# plt.legend(
#     title='지역명',
#     loc=(1.05 , 0.25)
# )

sns.move_legend(
    obj=fig,
    loc='upper right',
    bbox_to_anchor=(1.2, 1) # 범례의 생성 위치 지정
                            # 생성 위치를 기준으로 loc의 값으로 뒤집음
)
```

```python
g_local_year_price_mean = apt_price.groupby(['지역명', '연도']).분양가격.mean().astype(int).unstack()
fig = g_local_year_price_mean.plot(
    figsize=(10, 5),
    xticks=range(0,17)
)

# x축에 모든 지역명이 나오도록 생성
fig.get_xticklabels()
# fig.set_xticklabels(g_local_year_price_mean.index)
```

```python
sns.lineplot(
    data=apt_price,
    x='지역명',
    y='분양가격',
    hue='연도',
    errorbar=None
).figure.set_size_inches(10, 5)
```

```python
sns.barplot(
    data=apt_price,
    x='연도',
    y='분양가격',
    hue='지역명',
    errorbar=None
)
```

```python
# 카테고리 그래프
sns.catplot(
    data=apt_price,
    x='연도',
    y='분양가격',
    kind='point', # 특정 그래프만 사용가능
    col='지역명', # 카테고리를 분류할 컬럼명 
    col_wrap=4, # 한 라인에 들어올 그래프 수
    errorbar=None,
)
```

```python
sns.relplot(
    data=apt_price,
    x='연도',
    y='분양가격',
    kind='line', 
    col='지역명',
    col_wrap=4,
    errorbar=None
)
```

```python
sns.relplot(
    data=apt_price,
    x='연도',
    y='분양가격',
    kind='scatter', # 산점도
    col='지역명',
    col_wrap=4,
#     errorbar=None # 산점도는 오차막대 X
)
```

```python
# 연도별 평당분양가격
sns.boxplot(
    data=apt_price,
    x='연도',
    y='분양가격',
)
```
