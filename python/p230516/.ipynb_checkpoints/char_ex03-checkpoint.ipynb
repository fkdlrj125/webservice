```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame({
    'n1' : [1, 5, 10, 15, 20],
    'n2' : [1, 2, 3, 4, 20], 
    'n3' : [1, 17, 18, 19, 20] 
})
# 이상치, 극단값 : 데이터들 사이에 편차가 큰 데이터 
# 평균에 영향을 주기 때문에 처리해주는 것이 좋다
#  - 삭제, 그대로 사용, 값 변경, 평균으로 변경
# 이상치를 처리하기 전에 왜 이상치가 생겼는 지 파악 후 처리
# 편향된 데이터와 이상치로 정보가 왜곡되면, 전형적인 값 파악을 위해 중앙값 등 사용
df
```

```python
df.describe() # 중앙값과 평균의 차이가 크다면 데이터에 이상치가 있다고 판단 가능
              # 데이터가 편향됐다
```

```python
df.n1.hist(bins=3)
```

# kde

```python
# KDE(Kernel Density Estimation) : 커널 밀도 추정
sns.kdeplot(data=df.n1)
# 정규분포 그래프 
# 종 모양 : 이상적인, 정상적인 분포
```

```python
sns.kdeplot([1,2,3,4,5])
```

```python
sns.kdeplot(data=df.n2)
# 꼬리가 늘어진 그래프 : 오른쪽으로 편향된 분포(오른쪽에 이상치 존재)
# 평균값을 오른쪽으로 치우침
# 평균값이 대부분의 평균값보다 작음
```

```python
sns.kdeplot(data=df.n3)
# 꼬리가 늘어진 그래프 : 왼쪽으로 편향된 분포(왼쪽에 이상치 존재)
# 평균값을 왼쪽으로 치우침
# 평균값이 대부분의 평균값보다 큼
```

```python
df.describe().T
```

```python
# 오른쪽 편향, 왜도(skewness)
# Positive Skewness
sns.kdeplot(df.n2)
plt.grid()

# 평균 표시
plt.axvline(
    x=df.n2.mean(),
    color='r',
    linewidth=2,
    linestyle='--',
)

plt.axvline(
    x=df.n2.median(),
    color='g',
    linewidth=2,
    linestyle='--',
)
```

```python
# 왼쪽 편향, 왜도
# Negative Skewness
sns.kdeplot(data=df.n3)

plt.grid()

plt.axvline(
    x=df.n3.mean(),
    color='r',
    linestyle='--',
    linewidth=2
)

plt.axvline(
    x=df.n3.median(),
    color='g',
    linestyle='--',
    linewidth=2
)
```

```python
sns.kdeplot(data=df.n1)

plt.grid()

plt.axvline(
    x=df.n1.mean(),
    color='r',
    linewidth=2,
    linestyle='--'
)

plt.axvline(
    x=df.n1.median(),
    color='g',
    linewidth=2,
    linestyle='--'
)
```

```python
# 박스 플롯
# 데이터 세트에서 데이터가 분산되는 방식 설명
# 수염 : 데이터 확산
# 상자 : IQR() : 사분위 범위
```

# box

```python
# 데이터프레임을 통한 matplotlib메소드 
df.boxplot('n1')

# 박스 하단 : 1사분위
# 박스 안 초록선 : 2사분위(중앙값)
# 박스 상단 : 3사분위
# 상단 하단 선 : 1사분위 3사분위의 1.5배 값(상위, 하위 경계 / 이상치가 아닌 범위)
# 검은선 : 최대, 최소값
#  - 최대값과 최소값은 완전히 보장하진 못 함
# 아웃라이어 : 이상치, 극단값
```

```python
df.boxplot('n2')
# 이상치를 제외한 값들 중 상위 경계 내 값이 없기 때문에 최대값이 4로 표현
# 상위 경계를 벗어난 값은 이상치
```

```python
df.boxplot('n3')
# 이상치를 제외한 값들 중 하위 경계 내 값이 없기 때문에 최소값이 17로 표현
# 하위 경계를 벗어난 값은 이상치
```

```python
sns.boxplot(data=df)
```

```python
sns.violinplot(data=df.n1)
# 가운데 검은색 - 박스 그래프
# 주위 범위 - 데이터의 분포
```

```python
sns.violinplot(data=df)
# 한쪽으로 길게 나온 그래프는 해당 위치에 이상치 존재
```
