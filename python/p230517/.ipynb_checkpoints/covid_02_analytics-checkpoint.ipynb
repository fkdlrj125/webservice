```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rc('font', family='D2coding')
```

# csv 파일 읽어오기

```python
covid_df = pd.read_csv('corona_table.csv')
covid_df[:3]
```

```python
covid_df.info()
```

```python
covid_df[-2:]
```

```python
covid_df.확진자수.sum()
```

```python
# 확진자가 가장 많은 날
# covid_df[covid_df.확진자수==covid_df.확진자수.max()]
covid_df.sort_values(by='확진자수', ascending=False)[:1]
```

```python
# 사망자가 가장 많은 날
# covid_df[covid_df.사망자수==covid_df.사망자수.max()]
covid_df.sort_values(by='사망자수', ascending=False)[:1]
```

```python
# 일별 확진자 수 시각화
# 기준일의 Dtype이 object일 때
# 축에 오는 컬럼의 Dtype이면 텍스트가 겹쳐보임
date_decide = covid_df[['기준일', '확진자수']]
date_decide.plot(
    x='기준일',
    y='확진자수',
    legend=False
)
```

```python
covid_df.기준일 = covid_df.기준일.astype('datetime64[ns]')
covid_df.dtypes
```

```python
# 일별 확진자 수 시각화
# 기준일의 Dtype이 datetime일 때
# 텍스트가 겹치지 않게 회전이 생김
date_decide = covid_df[['기준일', '확진자수']]
date_decide.plot(
    x='기준일',
    y='확진자수',
    legend=False
)
```

```python
# 일별 확진자 수 시각화
# 기준일의 Dtype이 datetime일 때
date_decide = covid_df[['기준일', '확진자수']]
date_decide.plot(
    x='기준일',
    y='확진자수',
    figsize=(10,5),
    legend=False,
    rot=0,
    title='날짜별 확진자 수',
    xlabel='날짜',
    ylabel='확진자 수 (단위 : 명)',
    color='olive'
)

## x축 눈금 레이블 정렬
# ha = ['left', 'center', 'right']
plt.xticks(
    ha='center' # 레이블의 가운데가 눈금에 맞게 설정
)

plt.show()
```

```python
sns.lineplot(
    data=covid_df,
    x='기준일',
    y='확진자수',
)
```

```python
# 날짜별 사망자 수
date_death = covid_df[['기준일', '사망자수']]

date_death.plot(
    x='기준일',
    y='사망자수',
    figsize=(12,5),
    title='날짜별 사망자 수',
    color='orange',
    alpha=0.6,
    rot=0,
    xlabel='날짜',
    ylabel='확진자 수 (단위 : 명)',
    legend=False
)

plt.xticks(
    ha='center',
)

plt.show()
```

```python
plt.figure(
    figsize=(12,5)
)
sns.lineplot(
    data=covid_df,
    x='기준일',
    y='사망자수'
)
```

```python
# 날짜별 확진자 수와 사망자 수를 한 차트안에 시각화

plt.plot(date_death.기준일, date_death.사망자수)
plt.plot(date_decide.기준일, date_decide.확진자수)
# 데이터의 크기 차이가 심해 사망자 수가 제대로 표현 x
```

```python
covid_df[['기준일', '확진자수', '사망자수']].plot(
    x='기준일',
    figsize=(12,5),
    rot=0,
    color=['olive', 'orange'],
    alpha=0.6,
    title='날짜별 확진자 수와 사망자 수',
    secondary_y='사망자수', # 오른쪽에 2번째 y축 생성
    xlabel='날짜',
)
```

```python
help(covid_df.기준일.dt)
```

```python
# 월별 확진자 수와 사망자 수
covid_df_copy = covid_df
covid_df_copy.기준일 = covid_df_copy.기준일.dt.strftime('%Y-%m')
```

```python
g_date_ddsum = covid_df_copy.groupby('기준일')[['확진자수', '사망자수']].sum()

g_date_ddsum.plot(
    figsize=(15,5),
    color=['olive', 'orange'],
    title='월별 확진자 수와 사망자 수',
    secondary_y='사망자수',
    xticks=range(0,25),
    xlabel='월별',
    rot=20,
)

plt.show()
```
