# 결측치(결측값)
- 결측치(missing value)
    - NaN, Na, Null
    - 알려지지 않은 값, 알 수 없는 값
    

```python
import pandas as pd
import numpy
```

```python
data = {
    'city' : ['서울', '경기', '제주', '부산'],
    'total1' : [12000, 34000, 50500, 42000],
    'total2' : [12500, 38000, 150000, 37000]
}

# 데이터 프레임 생성
df = pd.DataFrame(data)
df
```

```python
df.dtypes
```

```python
df.loc[1,'total1'] = numpy.NAN
df
```

```python
df.dtypes
```

```python
# 결측치 확인 : info
df.info()
```

```python
# 결측치 확인 : 마스크
# 둘은 같은 함수
df.isnull()
df.isna()
```

```python
# 특정 컬럼에서 결측치 추출
# 결측치가 포함된 행을 확인 후 처리 방법을 선택하기 위해 추출
df[df.total1.isnull()]
df[df.total1.isna()]
```

```python
# 결측치를 원하는 값으로 변경
# fillna(value, axis, inplace)
df.fillna(0)
```

```python
# 결측치가 있는 행 또는 열을 삭제
# 모든 행이 결측치일 때 삭제하는 게 좋음
# DataFrame.dropna(axis, how, inplace)
df.dropna()
df.loc[len(df)] = numpy.NaN
df
```

```python
help(df.dropna)
```

```python
# 모든 값이 결측치인 행 삭제
df.dropna(how='all')
```
