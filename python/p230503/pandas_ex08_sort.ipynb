# 정렬
- 문법 : DataFrame.sort_values(by, axis, ascending, inplace)
    - by   : 정렬 기준이 될 행, 열
    - axis : 정렬이 될 축(default=0) -> 0 : 열정렬, 1 : 행정렬
    - ascending : 오름차순, 내림차순(default=True)
    - inplace : 원본 적용

```python
# 매직 명령어 - 현재 메모리에 있는 변수 출력
%who
```

```python
import pandas as pd

lt = [range(1,21,2), range(2,21,2), range(10,101,10)]
df = pd.DataFrame(data=lt, columns=[f'co{idx}' for idx in range(1,11)])
df
```

## 값 정렬

### 열 기준 : 오름차순(행정렬)
- 문법 : DataFrame.sort_values(by, inplace)

```python
help(df.sort_values)
```

```python
# co2 컬럼의 데이터를 기준으로 행정렬
df.sort_values('co2')
```

### 열 기준 : 내림차순(행정렬)
- 문법 : DataFrame.sort_values(by, ascending=False, inplace)

```python
df.sort_values('co2',ascending=False)
```

### 행 기준 : 오름차순(열정렬)
- 문법 : DataFrame.sort_values(by, axis=1, inplace)

```python
# 인덱스 1행의 데이터를 기준으로 열정렬
df.sort_values(1, axis=1)
```

### 행 기준 : 내림차순(열정렬)
- 문법 : DataFrame.sort_values(by, axis=1, ascending=False, inplace)

```python
df.sort_values(1, axis=1, ascending=False)
```

## 인덱스 정렬
- 문법 : DataFrame.sort_index(axis, ascending, key)
### 열 기준 : 오름차순(행정렬)
- 문법 : DataFrame.sort_index()

```python
df.sort_index()
```

### 열 기준 : 내림차순(행정렬)
- DataFrame.sort_index(ascending=False)

```python
df.sort_index(ascending=False)
```

### 행 기준 : 오름차순(열정렬)
- DataFrame.sort_index(axis=1)

```python
# 문자열 비교를 하기 때문에 co10이 앞으로 옴
df.sort_index(axis=1)
```

### 행 기준 : 내림차순(열정렬)
- DataFrame.sort_index(axis=1, ascending=False)

```python
df.sort_index(axis=1, ascending=False)
```

```python
help(df.sort_index)
```

```python
df = pd.DataFrame({"a": [1, 2, 3, 4]}, index=['A', 'b', 'C', 'd'])
df.sort_index(key=lambda x: x.str.upper())
```
