## 행, 열 삭제
- 문법 : DataFrame.drop(labels, axis, index, columns, inplace)
    - 복사본으로 작업
    - labels : 행의 라벨, 열의 라벨 
    - axis : 적용축 
        - 0 : index(default)
        - 1 : columns
    - index, columns : 라벨로 지정
    - inplace : 원본에 적용

```python
import pandas as pd
```

```python
lt = [range(1,21,2), range(2,21,2), range(10,101,10)]
df = pd.DataFrame(data=lt, columns=[f'co{idx}' for idx in range(1,11)])
df
```

```python
help(df.drop)
```

### 행 삭제
- 문법 : DataFrame.drop(labels or index)

#### labels로 삭제

```python
# 첫 번째 행 삭제
df.drop(0)
```

```python
# 첫 번째, 마지막 행 삭제
df.drop([0,2])
```

#### index로 삭제

```python
# 첫 번째, 마지막 행 삭제
df.drop(index=[0, 2])
```

### 열 삭제
- 문법 : DataFrame.drop(labels, axis=1 or columns)

#### labels로 삭제

```python
# co2 컬럼 삭제
df.drop('co2', axis=1)
```

```python
# co2, co6 컬럼 삭제
df.drop(['co2','co6'], axis=1)
```

#### columns로 삭제

```python
# co2 컬럼 삭제
df.drop(columns='co2')
```

```python
# co2, co6 컬럼 삭제
# df.drop(columns=['co2', 'co6'])
df.drop(columns=['co2', 'co6'])
```

### 원본 반영 삭제

```python
## co2 삭제
df.drop(columns='co2', inplace=True)
```

```python
df
```

### [참고] del 함수
- 문법 : del DataFrame[열]
- 원본 적용

```python
del df['co6']
```
