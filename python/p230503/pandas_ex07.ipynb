# 행 또는 열 추가

```python
import pandas as pd
```

```python
lt = [range(1,21,2), range(2,21,2), range(10,101,10)]
df = pd.DataFrame(data=lt, columns=[f'co{idx}' for idx in range(1,11)])
df
```

## 행 추가

### 마지막 행 추가
- 문법 : DataFrame.loc[추가할 라벨] = 데이터
- iloc는 현재 있는 index만 접근 가능

```python
df.loc[3] = [idx for idx in range(100, 110)]
```

```python
df.loc[5] = [idx for idx in range(200, 210)]
df
```

```python
# df.iloc[5] = [idx for idx in range(300,310)] # 위치 지정이 불가능해 Error
# df
```

```python
df.iloc[3] = [idx for idx in range(300,310)]
df
```

```python
# df.loc[100] = [500, 5001] # 컬럼의 수가 맞지 않아 Error
# df
```

```python
# 빈 행 추가
# df.loc[6] = None
# df.loc[7] = 0
# df.loc[8] = ''
# df.loc[9] = 1
df.loc[10] = pd.Series()
df
df.info()
```

## 열 추가

```python
df['co11'] = [1, 2, 3, 4, 5]
df
```

```python
# df.co12 = [1, 2, 3, 4, 5] # via를 통해서는 생성 불가
```

```python
# df['co12'] = pd.Series()
df['co13'] = 5
df
```
