```python
from urllib import request
from bs4 import BeautifulSoup
import requests
import pandas

# # 1. 선수 정보를 저장할 리스트 생성(딕셔너리)
# number = []
# name = []
# position = []
# age = []
# nation = []
# team = []
# value = []

# 1-1. 선수 정보를 저장할 리스트 생성(리스트)
player_list = list()

headers = {'user-agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/112.0.0.0 Safari/537.36'}

for page in range(1,5):
    # 2. 웹 페이지 요청
    url = f'https://www.transfermarkt.com/spieler-statistik/wertvollstespieler/marktwertetop?page={page}'

    html = requests.get(url, headers=headers)
    print(html.status_code)

    # 3. 파싱
    soup = BeautifulSoup(html.text, 'html.parser')

    # 4. 엘리먼트 추출
    # 값을 Iterable로 넣어줘 여러 개 비교 가능
    player_info = soup.find_all('tr', class_=['odd','even'])

    # 5. 리스트에 선수 정보 저장
    for info in player_info:
##        딕셔너리 형태
#         player = info.find_all('td')
#         number.append(player[0].text)
#         name.append(player[3].text)
#         position.append(player[4].text)
#         age.append(player[5].text)
#         nation.append(player[6].img['alt'])
#         team.append(player[7].a['title'])
#         value.append(player[8].a.text)

#       리스트 형태
        player = info.find_all('td')
        number = player[0].text
        name = player[3].text
        position = player[4].text
        age = player[5].text
        nation = player[6].img['alt']
        team = player[7].a['title']
        value = player[8].a.text.replace('€', '').split('.')[0]
        player_list.append([number, name, position, age, nation, team, value])

# # 6 DataFrame 생성(딕셔너리)
# df = pandas.DataFrame({
#     'number' : number,
#     'name' : name,
#     'position' : position,
#     'age' : age,
#     'nation' : nation,
#     'team' : team,
#     'value' : value
# })

# 6-1 DataFrame 생성(리스트)
df = pandas.DataFrame(player_list, columns=['number', 'name', 'position', 'age', 'nation', 'team', 'value'])

# 7. csv 파일 저장 - notebook의 상대 경로 : 현재 디렉토리
df.to_csv('market100.csv', index=False)
print('** 저장 완료 **')
```
