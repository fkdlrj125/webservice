## u.data : 영화 평점
- 사용자 아이디, 영화 아이디, timestamp
- itmestamp : 연도/ 날짜 /시간


```python
data_col = ['user_id', 'movie_id', 'rating', 'timestamp'] 

ratings = pd.read_csv('u.data', sep='\t', names=data_col)
ratings

ratings.set_index('user_id', inplace=True)

ratings.sort_index(inplace=True)
ratings[:3]
```

```python
ratings.to_csv('ratings.csv', encoding='utf-8')
```

# 추천 방식
## best-seller(인기 상품)
- 가장 인기있는 제품(상품)을 추천
- 사용자 개인(개별) 정보가 없을 때
- 각 영화의 평점의 평균을 구한 다음, 평균값이 높은 것을 순서대로 추천

```python
# 영화별 평점의 평균
g_movie_rating_mean = ratings.groupby('movie_id').rating.mean()
g_movie_rating_mean[:3]
```

```python
g_movie_rating_mean.sort_values(ascending=False)[:20]
```

```python
# Best-seller 추천
# 함수 정의
def recom_movie(n_items):
    movie_sort = g_movie_rating_mean.sort_values(ascending=False)[:n_items]
    recommend = movies.loc[movie_sort.index, 'title']
    return recommend
```

```python
# 상위 15개의 영화 추천
recommend = recom_movie(15)
recommend
```
