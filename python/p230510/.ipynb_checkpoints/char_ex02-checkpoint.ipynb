```python
# 한글 폰트 설정 - 해당 pc의 폰트 적용
# plt.rc('font', family='Malgun Gothic')
# plt.rc('axes', unicode_minus=False)
# 설정을 딕셔너리로 저장하고 있어 딕셔너리 키 값을 통해 설정 가능
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

# data
data = np.arange(0, 10, 0.01)

sin_y = np.sin(data)
cos_y = np.cos(data)

# 차트 타이틀 설정
plt.title('sin cos 그래프')

# 선형 그래프 생성
plt.plot(data, sin_y, label='sin')
plt.plot(data, cos_y, label='cos')

# 범례 출력
plt.legend()

# 그리드 설정 : 차트에 격자 설정
plt.grid()

# x축 이름 설정
plt.xlabel('x축')

# y축 이름 설정
plt.ylabel('y축')

plt.show()
```

```python
# 원형 그래프

# data
data = [10, 20, 30, 5]

# 원형 그래프 생성
plt.pie(data, 
        labels=['하나', '둘', '셋', '넷'],  # 조각의 이름
        autopct='%.1f%%', # 데이터의 비율 
        startangle=45, # 시작 각도
#         counterclock=False, # 조각 생성 방향
        explode=[0.25,0,0,0],
        shadow=True,
#         colors=['#ff9999', '#ffc000', '#8fd9b6', '#abcdef'],
        colors=['gold', 'silver', 'whitesmoke', 'lightgray'],
        wedgeprops={'width' : 0.8, 'edgecolor' : 'w', 'linewidth' : 1}
       )

# 차트 실행
plt.show()
```

```python
# 원형 그래프
# << 코드 정리 >>

# data
data = [10, 20, 30, 5]
labels=['하나', '둘', '셋', '넷']
explode=[0.25,0,0,0]
#colors=['#ff9999', '#ffc000', '#8fd9b6', '#abcdef'], 
colors=['gold', 'silver', 'whitesmoke', 'lightgray']
wedgeprops={'width' : 0.8, 'edgecolor' : 'w', 'linewidth' : 1}
# - width : 반지름 비율
# - edgecolor : 테두리 색
# - linewidth : 테두리 두께

# 원형 그래프 생성
plt.pie(data, 
        labels=labels,  # 조각의 이름
        autopct='%.1f%%', # 데이터의 비율 
        startangle=45, # 시작 각도
#         counterclock=False, # 조각 생성 방향
        explode=explode, # 중심에서 조각 분리
        shadow=True,
        colors=colors, # 조각 색 지정
        wedgeprops=wedgeprops # 조각 설정
       )

# 차트 실행
plt.show()
```

```python
# 데이터 1000개 생성
data1 = np.random.randn(1000)
data2 = np.random.randn(1000) + 1

# 히스토그램 그래프 생성 - bar
plt.figure() # 차트 생성
plt.title('histtype - bar')
plt.hist([data1, data2], bins=30, label=['data1', 'data2'], histtype='bar')
plt.legend()

# barstacked
plt.figure() # 차트 생성
plt.title('histtype - barstacked')
plt.hist([data1, data2], bins=30, label=['data1', 'data2'], histtype='barstacked')
plt.legend()

# step
plt.figure() # 차트 생성
plt.title('histtype - step')
plt.hist([data1, data2], bins=30, label=['data1', 'data2'], histtype='step')
plt.legend()

# stepfilled
plt.figure() # 차트 생성
plt.title('histtype - stepfilled')
plt.hist([data1, data2], bins=30, label=['data1', 'data2'], histtype='stepfilled')

plt.legend()
plt.show()
```

```python
# subplot 방법 1
# 데이터 1000개 생성
data1 = np.random.randn(1000)
data2 = np.random.randn(1000) + 1

# fig = plt.figure() # 차트 생성
# ax1 = fig.add_subplot(2, 2, 1)
# ax2 = fig.add_subplot(2, 2, 2)
# ax3 = fig.add_subplot(2, 2, 3)
# ax4 = fig.add_subplot(2, 2, 4)

# 히스토그램 그래프 생성 - bar
plt.subplot(2, 2, 1)
plt.hist([data1, data2], bins=30, label=['data1', 'data2'], histtype='bar')
plt.title('histtype - bar')

# barstacked
plt.subplot(2, 2, 2)
plt.hist([data1, data2], bins=30, label=['data1', 'data2'], histtype='barstacked')
plt.title('histtype - barstacked')

# step
plt.subplot(2, 2, 3)
plt.hist([data1, data2], bins=30, label=['data1', 'data2'], histtype='step')
plt.title('histtype - step')

# stepfilled
plt.subplot(2, 2, 4)
plt.hist([data1, data2], bins=30, label=['data1', 'data2'], histtype='stepfilled')
plt.title('histtype - stepfilled')


plt.tight_layout()

plt.show()
```

```python
# subplot 방법 2
# 데이터 1000개 생성
data1 = np.random.randn(1000)
data2 = np.random.randn(1000) + 1

# fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)
fig, ax = plt.subplots(2, 2)
fig.subplots_adjust(hspace=0.5) # 수평선 간격 조절
fig.set_size_inches(10, 8) # 

# 히스토그램 그래프 생성 - bar
# ax1.hist([data1, data2], bins=30, label=['data1', 'data2'], histtype='bar') # 각 위치를 저장한 변수 이용
# ax1.set_title('histtype - bar')
ax[0][0].hist([data1, data2], bins=30, label=['data1', 'data2'], histtype='bar') # 모든 위치를 저장한 변수의 인덱스를 이용(선호)
ax[0][0].set_title('histtype - bar')

# barstacked
# ax2.hist([data1, data2], bins=30, label=['data1', 'data2'], histtype='barstacked')
# ax2.set_title('histtype - barstacked')
ax[0][1].hist([data1, data2], bins=30, label=['data1', 'data2'], histtype='barstacked')
ax[0][1].set_title('histtype - barstacked')

# step
# ax3.hist([data1, data2], bins=30, label=['data1', 'data2'], histtype='step')
# ax3.set_title('histtype - step')
ax[1][0].hist([data1, data2], bins=30, label=['data1', 'data2'], histtype='step')
ax[1][0].set_title('histtype - step')

# stepfilled
# ax4.hist([data1, data2], bins=30, label=['data1', 'data2'], histtype='stepfilled')
# ax4.set_title('histtype - stepfilled')
ax[1][1].hist([data1, data2], bins=30, label=['data1', 'data2'], histtype='stepfilled')
ax[1][1].set_title('histtype - stepfilled')

plt.show()
```

```python
# x축 
x = np.arange(0, 100)

# y축
y1 = [v*v for v in x]
y2 = [-v for v in x]
y3 = [v*v/2 for v in x]

# 그래프 생성
# 차트 1개로 그래프 2개 출력
# plt.plot(x, y1)
# plt.plot(x, y2)

# 차트 2개로 그래프 출력
# plt.figure()
# plt.plot(x, y1)
# plt.figure()
# plt.plot(x, y2)

# subplot 방법 1
# 차트 1개의 구역을 나눠 그래프 2개 출력
fig = plt.figure()
ax1 = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 1, 2)

ax1.plot(x, y1, x, y3)
ax2.plot(x, y2)

plt.tight_layout(pad=1.5)

# 차트 실행
plt.show()
```

```python
# subplot 방법 2
# x축 
x = np.arange(0, 100)

# y축
y1 = [v*v for v in x]
y2 = [-v for v in x]
y3 = [v*v/2 for v in x]

# 그래프 생성
fig, (ax1, ax2) = plt.subplots(2, 1) # subplot 생성
fig.subplots_adjust(hspace=0.5) # subplot 간격 조정
fig.set_size_inches(10, 8) # 차트 크기 변경(인치)

# 첫 번째 subplot
ax1.plot(x, y1, x , y3)
ax1.set_xlabel('X 축')
ax1.set_ylabel('y 축')
ax1.grid()

# 두 번째 subplot
ax2.plot(x, y2)
ax2.set_xlabel('X 축')
ax2.set_ylabel('y 축')
ax2.grid()

# 차트 실행
plt.show()
```

```python
# data
one = [10, 20, 30, 40] 
two = [40, 30, 20, 10]

# 제목
plt.title('color')

# 그래프 생성
# plt.plot(one, color='DarkRed', label='one', linestyle='--')
# plt.plot(two, 'pink', label='two', linestyle=':')

plt.plot(one, 'b--', label='one') # 배이직컬러는 라인스타일도 같이 설정 가능
plt.plot(two, 'r:', label='two')

# 범례 표시
plt.legend()

plt.show()
```

```python
# data
one = [10, 20, 30, 40] 
two = [40, 30, 20, 10]

# 제목
plt.title('color')

# 그래프 생성
plt.plot(one, 'b.', label='one')
plt.plot(two, 'r^', label='two')

# 범례 표시
plt.legend(loc=10)

plt.show()
```
