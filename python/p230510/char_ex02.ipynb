```python
#  1000 
data1 = np.random.randn(1000)
data2 = np.random.randn(1000) + 1

#    - bar
plt.figure() #  
plt.title('histtype - bar')
plt.hist([data1, data2], bins=30, label=['data1', 'data2'], histtype='bar')
plt.legend()

# barstacked
plt.figure() #  
plt.title('histtype - barstacked')
plt.hist([data1, data2], bins=30, label=['data1', 'data2'], histtype='barstacked')
plt.legend()

# step
plt.figure() #  
plt.title('histtype - step')
plt.hist([data1, data2], bins=30, label=['data1', 'data2'], histtype='step')
plt.legend()

# stepfilled
plt.figure() #  
plt.title('histtype - stepfilled')
plt.hist([data1, data2], bins=30, label=['data1', 'data2'], histtype='stepfilled')

plt.legend()
plt.show()
```

```python
# subplot  1
#  1000 
data1 = np.random.randn(1000)
data2 = np.random.randn(1000) + 1

# fig = plt.figure() #  
# ax1 = fig.add_subplot(2, 2, 1)
# ax2 = fig.add_subplot(2, 2, 2)
# ax3 = fig.add_subplot(2, 2, 3)
# ax4 = fig.add_subplot(2, 2, 4)

#    - bar
plt.subplot(2, 2, 1)
plt.hist([data1, data2], bins=30, label=['data1', 'data2'], histtype='bar')
plt.title('histtype - bar')

# barstacked
plt.subplot(2, 2, 2)
plt.hist([data1, data2], bins=30, label=['data1', 'data2'], histtype='barstacked')
plt.title('histtype - barstacked')

# step
plt.subplot(2, 2, 3)
plt.hist([data1, data2], bins=30, label=['data1', 'data2'], histtype='step')
plt.title('histtype - step')

# stepfilled
plt.subplot(2, 2, 4)
plt.hist([data1, data2], bins=30, label=['data1', 'data2'], histtype='stepfilled')
plt.title('histtype - stepfilled')


plt.tight_layout()

plt.show()
```

```python
# subplot  2
# x 
x = np.arange(0, 100)

# y
y1 = [v*v for v in x]
y2 = [-v for v in x]
y3 = [v*v/2 for v in x]

#  
fig, (ax1, ax2) = plt.subplots(2, 1) # subplot 
fig.subplots_adjust(hspace=0.5) # subplot  
fig.set_size_inches(10, 8) #   ()

#   subplot
ax1.plot(x, y1, x , y3)
ax1.set_xlabel('X ')
ax1.set_ylabel('y ')
ax1.grid()

#   subplot
ax2.plot(x, y2)
ax2.set_xlabel('X ')
ax2.set_ylabel('y ')
ax2.grid()

#  
plt.show()
```
